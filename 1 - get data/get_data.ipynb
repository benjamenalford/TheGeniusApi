```python
import requests
import pandas as pd
import pymongo
import json
from config import genius_token
```

```python
artist_id = "360472"
page =1
next_page = None
url = f"https://api.genius.com/artists/{artist_id}/songs?per_page=50=&page={page}"

headers = {
    "Authorization": genius_token
}

response = requests.get(url, headers=headers)

response_json = response.json()
next_page = response_json['response']['next_page']
```

```python
songList_df = pd.DataFrame(response_json['response']['songs'])
```

```python
songList_df = songList_df[['id', 'artist_names', 'full_title','lyrics_state', 'path',
       'release_date_for_display', 'title','url','header_image_thumbnail_url', 'header_image_url' ]]
```

```python
#
serverUrl = "mongodb://localhost:27017"
client = pymongo.MongoClient(serverUrl)
db = client.genius
song_collection = db.songs
```
